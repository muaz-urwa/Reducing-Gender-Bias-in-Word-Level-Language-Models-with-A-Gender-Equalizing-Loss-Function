```python
def tsne_plot(model,words,corpus):
    "Creates and TSNE model and plots it"
    labels = []
    tokens = []
    
    words = [w for w in words if w in corpus.words2idx]
    words_idx = torch.tensor([corpus.words2idx[w] for w in words]).to(device)
    tokens = model.encoder(words_idx).cpu().detach().numpy()
    labels = words 
    
    tsne_model = TSNE(perplexity=40, n_components=2, init='pca', n_iter=2500, random_state=23)
    new_values = tsne_model.fit_transform(tokens)

    x = []
    y = []
    for value in new_values:
        x.append(value[0])
        y.append(value[1])
        
    plt.figure(figsize=(10, 6)) 
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(labels[i],
                     xy=(x[i], y[i]),
                     xytext=(5, 2),
                     textcoords='offset points',
                     ha='right',
                     va='bottom')
    plt.show()
```

```python
def load_doc(filename):
    file = open(filename, 'r', encoding='utf-8-sig')
    text = file.read()
    file.close()
    return text
```

```python
def getUpper(word):
    #print(word)
    word = word[0].upper() + word[1:]
    return word
```

```python
def distance(v1,v2):
    return torch.pow(torch.sum(torch.pow(v1-v2,2)),0.5).item()
```

```python
def getParameters(model):
    # load vocab & file index  
    if 'cda' in model:
        path = 'processedDataCda/'
        vocab = preprocess.read_vocab(os.path.join('data','VOCAB_cda.txt'))
        idx_train = pd.read_json('data/idx_train_cda.json')
        idx_val = pd.read_json('data/idx_val_cda.json')
        idx_test = pd.read_json('data/idx_test_cda.json')
    else:
        path = 'processedData/'
        vocab = preprocess.read_vocab(os.path.join('data','VOCAB.txt'))
        idx_train = pd.read_json('data/idx_train.json')
        idx_val = pd.read_json('data/idx_val.json')
        idx_test = pd.read_json('data/idx_test.json')
            
    vocab.append('<eos>')
    words2idx = {item : index for index, item in enumerate(vocab)}
           
    return path, vocab, words2idx, idx_train, idx_val, idx_test
```

```python
femaleFile = '/home/urwa/Documents/Courses/NLU/GBLM/GBLM/Loss Function/gender_words/female_word_file.txt'
maleFile = '/home/urwa/Documents/Courses/NLU/GBLM/GBLM/Loss Function/gender_words/male_word_file.txt'

def getGenderIdx(femaleFile, maleFile, word2idx):
    female_word_list = load_doc(femaleFile).split('\n')
    male_word_list = load_doc(maleFile).split('\n')
    #print(len(female_word_list),len(male_word_list))
    pairs = [ (word2idx[f],word2idx[m]) for f,m in zip(female_word_list,male_word_list) \
             if f in word2idx and m in word2idx]
    femaleIdx = [ f for f,m in pairs]
    maleIdx = [ m for f,m in pairs]
    #print(len(femaleIdx),len(maleIdx))
    return femaleIdx,maleIdx
```

```python
def getBias(model):
    femaleWords,maleWords = getGenderIdx(femaleFile, maleFile, corpus.words2idx)
    femaleWords = torch.tensor(femaleWords).to(device)
    maleWords = torch.tensor(maleWords).to(device)
    occ = [w for w in Occupations if w in corpus.words2idx]
    occupations_indexes = torch.tensor([corpus.words2idx[w] for w in occ ]).to(device)
    
    
    with torch.no_grad():
        occ_word_vectors = model.encoder(occupations_indexes)
        male_vectors = model.encoder(maleWords)
        female_vectors = model.encoder(femaleWords)
    avgBlist = []
    avgModBlist = []
    for o in  occ_word_vectors:
        bList = []
        for i in range(len(male_vectors)):
            b = distance(o,male_vectors[i])-distance(o,female_vectors[i])
            bList.append(b)
        avgBlist.append(np.mean(np.array(bList)))
        avgModBlist.append(np.mean(np.abs(np.array(bList))))
    #print(list(zip(Occupations,avgModBlist)))
    return avgBlist,avgModBlist,occ# np.mean(avgBlist),np.mean(avgModBlist)
```

```python
import operator
def load_glove(glove_file, n_vecs=20000):
    """ """
    tok2vec = {}
    with open(glove_file, 'r') as glove_fh:
        for i, row in enumerate(glove_fh):
            word, vec = row.split(' ', 1)
            tok2vec[word] = np.array([float(n) for n in vec.split(' ')])
            #if i >= n_vecs:
            #    break
    return tok2vec
```

```python
def cpudistance(v1,v2):
    return np.power(np.sum(np.power(v1-v2,2)),0.5)
```

```python
def getBiasGlove(glove_vecs):
    femaleWords = load_doc(femaleFile).split('\n')
    maleWords = load_doc(maleFile).split('\n')
    
    femaleWords = [w for w in femaleWords if len(w)>0]
    maleWords = [w for w in maleWords if len(w)>0]
    pairs = [ (f,m) for f,m in zip(femaleWords,maleWords) \
             if f in glove_vecs and m in glove_vecs]
    femaleWords = [ f for f,m in pairs]
    maleWords = [ m for f,m in pairs]
    
    #femaleWords,maleWords = getGenderIdx(femaleFile, maleFile, corpus.words2idx)
    #femaleWords = torch.tensor(femaleWords).to(device)
    #maleWords = torch.tensor(maleWords).to(device)
    occ = [w for w in Occupations if w in glove_vecs]
    #occupations_indexes = torch.tensor([corpus.words2idx[w] for w in occ ]).to(device)
    #print(maleWords)

    occ_word_vectors = [glove_vecs[o] for o in occ] 
    male_vectors = [glove_vecs[o] for o in maleWords] 
    female_vectors = [glove_vecs[o] for o in femaleWords] 
    
    avgBlist = []
    avgModBlist = []
    for o in  occ_word_vectors:
        bList = []
        for i in range(len(male_vectors)):
            b = cpudistance(o,male_vectors[i])-cpudistance(o,female_vectors[i])
            bList.append(b)
        avgBlist.append(np.mean(np.array(bList)))
        avgModBlist.append(np.mean(np.abs(np.array(bList))))
    #print(list(zip(Occupations,avgModBlist)))
    return avgBlist,avgModBlist,occ# np.mean(avgBlist),np.mean(avgModBlist)
```
