```python
import os
```

```python
# load doc into memory
def load_doc(filename):
    # open the file as read only
    file = open(filename, 'r', encoding='utf-8-sig')
    # read all text
    text = file.read()
    # close the file
    file.close()
    return text
```

```python
output_data_dir = './Bias/'
txtDir = './generated/'
models = os.listdir(txtDir)
models
```

```python
mentionsList = []
for m in models:
    data ={}
    print("-----------------------------")
    print("Processing: ",m)
    print("Files:",len(os.listdir(txtDir + m)))
    maleCount = 0
    femaleCount = 0
    mention = []
    for file in os.listdir(txtDir + m):
        txt_path = txtDir + m + '/' + file
        lines = load_doc(txt_path).split('.')
        for l in lines:
            if 'pregnant' in l or 'Pregnant' in l:
                mention.append(l)
                #print(l)
                #print()
    mentionsList.append(mention)
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
index = 5
print(models[index])
print(len(mentionsList[index]))
print('------------------  Wrong Use  ----------------------------')
for line in mentionsList[index]:
    doc = nlp(line)
    for token in doc:
        children = [child.text for child in token.children]
        if 'pregnant' in children and ('he' in children or 'He' in children) :
            print(token.text, token.dep_, token.head.text, token.head.pos_,
                    [child for child in token.children])
            print('sentence:  ',line)
    print('--------------')
print('------------------  Right Use  ----------------------------')
for line in mentionsList[index]:
    doc = nlp(line)
    for token in doc:
        children = [child.text for child in token.children]
        if 'pregnant' in children and ('she' in children or 'She' in children):
            print(token.text, token.dep_, token.head.text, token.head.pos_,
                    [child for child in token.children])
            print('sentence:  ',line)
    print('--------------')
```
